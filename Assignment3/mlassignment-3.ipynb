```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('admission - admission.csv')
```

```python
#scatter plot of GRE Score vs chances of admit
plt.scatter(data['GRE Score'],data['Chance of Admit'],color='red')
plt.xlabel('GRE Score')
plt.ylabel('Chance of admit')
plt.show()
```

```python
#bar plot of university rating vs CGPA
plt.bar(data['University Rating'],data['CGPA'],color='green')
plt.xlabel('University Rating')
plt.ylabel('CGPA')
plt.show()
```

```python
#box plot
sns.boxplot(x=data['Research'], y=data['TOEFL Score'])
plt.show()
```

```python
#Pie chart of university ratings
plt.pie(data['University Rating'].value_counts(), labels=data['University Rating'].unique(), autopct='%1.1f%%')
plt.show()
```

```python
#line chart for CGPA vs chance of admit
data_sorted = data.sort_values(by='CGPA')
plt.figure(figsize=(10, 6))
plt.plot(data_sorted['CGPA'], data_sorted['Chance of Admit'], marker='o', linestyle='-', color='b')

plt.title('Chance of Admit vs CGPA', fontsize=16)
plt.xlabel('CGPA', fontsize=14)
plt.ylabel('Chance of Admit', fontsize=14)
```
